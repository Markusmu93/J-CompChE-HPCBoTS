```python
import numpy as np
from scipy.optimize import direct, Bounds

# Define constants for the Hartmann 6D function
alpha = np.array([1.0, 1.2, 3.0, 3.2])
A = np.array([
    [10, 3, 17, 3.5, 1.7, 8],
    [0.05, 10, 17, 0.1, 8, 14],
    [3, 3.5, 1.7, 10, 17, 8],
    [17, 8, 0.05, 10, 0.1, 14]
])
P = 1e-4 * np.array([
    [1312, 1696, 5569, 124, 8283, 5886],
    [2329, 4135, 8307, 3736, 1004, 9991],
    [2348, 1451, 3522, 2883, 3047, 6650],
    [4047, 8828, 8732, 5743, 1091, 381]
])

def hartmann_6d(X):
    """6D Hartmann function modified to accept both 1D and 2D array input."""
    X = np.atleast_2d(X)  # Ensure X is 2D for consistent numpy operations
    
    # Calculate the Hartmann value for each row/sample
    result = np.zeros(X.shape[0])
    for i in range(4):
        result -= alpha[i] * np.exp(-np.sum(A[i] * (X - P[i])**2, axis=1))
    
    return -1*np.squeeze(result)  # Ensure to return a scalar if a single value

# Bounds for each of the 6 dimensions, typically in the range [0, 1] for the Hartmann function
bounds = Bounds([0] * 6, [1] * 6)

# Using scipy.optimize.direct to find the minimum of the Hartmann function
# result = direct(hartmann_6d, bounds, maxiter=10000)   # but here we would look for the maximum !

# Print the results
# print("Optimal parameters:", result.x)
# print("Function value at optimal parameters:", result.fun)

# Known optimum
print("Known value at the optimum = ", hartmann_6d([0.20169,0.150011,0.476874, 0.275332,0.311652,0.6573]))

global_maximum_value = 3.32237   # from literature : https://www.sfu.ca/~ssurjano/hart6.html

print("\n New optvalue for comparison in log regret: ", global_maximum_value)
```

```python
import pickle
import numpy as np
import os

# Known global minimum for the Hartmann 6D function
global_minimum_value = -3.32237

# Directory where the pickle files are stored
directory_path = './results/'

# Function to load pickled data
def load_pickle(file_path):
    with open(file_path, "rb") as file:
        data = pickle.load(file)
    return data

# Function to calculate regret
def calculate_regret(histories, global_min):
    regrets = []
    for history in histories:
        # Find the best value within each batch
        best_values = np.array([np.max(batch) for batch in history])
        # Calculate the regret for each iteration
        run_regret = global_min - best_values
        regrets.append(run_regret)
    # Convert the list of regrets to a 2D NumPy array for easier processing
    regrets = np.array(regrets)
    # Calculate mean and standard deviation of regret at each iteration across all runs
    mean_regret = np.mean(regrets, axis=0)
    std_regret = np.std(regrets, axis=0)
    return mean_regret, std_regret

# Process each history file and calculate the regret
mean_regrets = []
std_regrets = []
method_names = []

# Get all the history files
# history_files = [f for f in os.listdir(directory_path) if f.startswith("histories_dict_function_experimental")]
history_files = ['histories_dict_function_experimental_TS-Constrained_EI_hartmann6d.pkl', 'histories_dict_function_experimental_Random_hartmann6d.pkl']
# Process each file
for history_file in history_files:
    file_path = os.path.join(directory_path, history_file)
    histories_dict = load_pickle(file_path)
    
    # There should be only one method per file, so we extract that method's name and its histories
    method_name, histories = next(iter(histories_dict.items()))
    mean_regret, std_regret = calculate_regret(histories, global_minimum_value)
    
    mean_regrets.append(mean_regret)
    std_regrets.append(std_regret)
    method_names.append(method_name)
```

```python
def log_transform(data):
    """Apply a logarithmic transformation to the data while preserving the sign."""
    return np.sign(data) * np.log(np.abs(data) + 1)

def visualize_optimization_convergence_log_transformed(histories_dict, title):
    # Create a figure and a set of subplots
    fig, ax = plt.subplots()
    
    # For each history in the dictionary
    for label, histories in histories_dict.items():
        if label == 'TS-Constrained_PI':
            pass
        else: 
            # Calculate the best result found so far at each iteration for each run
            best_results_per_run = [np.maximum.accumulate([np.max(y) for y in history]) for history in histories]

            # Apply log transformation
            best_results_per_run_transformed = [log_transform(results) for results in best_results_per_run]
            
            # Calculate the mean and percentiles of the transformed best results at each iteration
            mean_best_results_transformed = np.mean(best_results_per_run_transformed, axis=0)
            percentile_10_transformed = np.percentile(best_results_per_run_transformed, 10, axis=0)
            percentile_90_transformed = np.percentile(best_results_per_run_transformed, 90, axis=0)
            
            # Plot the mean of the transformed best results
            ax.plot(mean_best_results_transformed, label=f'Mean best result ({label})')
            
            # Fill the area between the transformed percentiles
            ax.fill_between(range(len(mean_best_results_transformed)), percentile_10_transformed, percentile_90_transformed, alpha=0.2)
        
    # Set the title and labels
    ax.set_title(title)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Best result (Log Transformed)')
    ax.set_ylim([-5,5])
    # Add a legend
    ax.legend(bbox_to_anchor=(1.04,1))
    
    # Show the plot
    plt.show()

# Loop through each file in the results directory
for history_filename in os.listdir('./results'):
    # Extract the number of unconstrained dimensions from the filename
    num_unconstrained = int(history_filename.split('_')[-1].split('.pkl')[0])
    
    # Load the histories for the current number of unconstrained dimensions
    histories = load_histories(f'./results/histories_dict_num_unconstrained_{num_unconstrained}.pkl')
    
    #print(histories)
    # Count the number of constrained dimensions (assuming a 4D domain)
    num_constrained = 4 - num_unconstrained
    # Visualize the optimization convergence with log transformation
    visualize_optimization_convergence_log_transformed(histories, 'Number of Constrained Dimensions: 3 (Log Transformed)')
```

```python
def rosenbrock_4d_modified(X):
    """4D Rosenbrock function modified to accept both 1D and 2D array input."""
    # Ensure input is a numpy array
    X = np.asarray(X)
    
    # If 1D input, reshape to 2D
    if X.ndim == 1:
        X = X.reshape(1, -1)
    
    # Calculate Rosenbrock value for each row/sample
    values = [sum(100.0 * (x[i+1] - x[i]**2.0)**2.0 + (1 - x[i])**2.0 for i in range(3)) for x in X]
    
    return -1 * np.array(values)


# Define Domain
domain = [{'name': 'x1', 'type': 'constrained', 'domain':(-2,2)},
        {'name': 'x2', 'type': 'constrained', 'domain':(-2,2)},
        {'name': 'x3', 'type': 'unconstrained', 'domain':(-2,2)},
        {'name': 'x4', 'type': 'unconstrained', 'domain':(-2,2)}]

bo_test_gp_experiment = BO_TS_constrained_direct.pc_BO_TS(objective_function=rosenbrock_4d_modified, domain=domain, B=4, n_init=1, kernel=None, acquisition_type='UCB')
bo_test_gp_experiment.optimize(20)
```

```python
def hartmann_6d(X):
    """6D Hartmann function"""
    # Ensure input is a numpy array
    X = np.asarray(X)
    
    # If 1D input, reshape to 2D
    if X.ndim == 1:
        X = X.reshape(1, -1)
    
    # Constants for the Hartmann 6D function
    alpha = np.array([1.0, 1.2, 3.0, 3.2])
    A = np.array([[10, 3, 17, 3.50, 1.7, 8],
                  [0.05, 10, 17, 0.1, 8, 14],
                  [3, 3.5, 1.7, 10, 17, 8],
                  [17, 8, 0.05, 10, 0.1, 14]])
    P = 0.0001 * np.array([[1312, 1696, 5569, 124, 8283, 5886],
                           [2329, 4135, 8307, 3736, 1004, 9991],
                           [2348, 1451, 3522, 2883, 3047, 6650],
                           [4047, 8828, 8732, 5743, 1091, 381]])
    
    # Calculate Hartmann value for each row/sample
    values = []
    for x in X:
        external_sum = 0
        for i in range(4):
            internal_sum = 0
            for j in range(6):
                internal_sum += A[i, j] * (x[j] - P[i, j])**2
            external_sum += alpha[i] * np.exp(-internal_sum)
        values.append(-external_sum)
    
    return -1 * np.array(values)



# Define Domain
domain = [{'name': 'x1', 'type': 'constrained', 'domain':(0,1)},
        {'name': 'x2', 'type': 'constrained', 'domain':(0,1)},
        {'name': 'x3', 'type': 'constrained', 'domain':(0,1)},
        {'name': 'x4', 'type': 'unconstrained', 'domain':(0,1)},
        {'name': 'x5', 'type': 'unconstrained', 'domain':(0,1)},
        {'name': 'x6', 'type': 'unconstrained', 'domain':(0,1)}]

bo_test_gp_experiment = BO_PC_Nested_direct_gpucb.pc_BO_nested(objective_function=hartmann_6d, domain=domain, B=4, n_init=1, kernel=None, acquisition_type='GP-UCB')
bo_test_gp_experiment.optimize(20)
```

```python
from skopt import gp_minimize
from skopt.space import Real
from skopt.utils import use_named_args
import numpy as np
import matplotlib.pyplot as plt

# Negated Rosenbrock function for maximization
def neg_rosenbrock_6d(x):
    sum = 0
    for i in range(5):  # For 6D, we iterate through the first 5 elements
        sum += 100 * (x[i+1] - x[i]**2)**2 + (1 - x[i])**2
    return -sum  # Negate for maximization

# Define the space of parameters
space  = [Real(-2.048, 2.048, name='x1'),
          Real(-2.048, 2.048, name='x2'),
          Real(-2.048, 2.048, name='x3'),
          Real(-2.048, 2.048, name='x4'),
          Real(-2.048, 2.048, name='x5'),
          Real(-2.048, 2.048, name='x6')]

# Use the decorator to convert the function to accept a list of named arguments
@use_named_args(space)
def objective_function(**params):
    x = [params['x1'], params['x2'], params['x3'], params['x4'], params['x5'], params['x6']]
    return neg_rosenbrock_6d(x)

# Perform Bayesian Optimization
res_gp = gp_minimize(objective_function,  # the function to minimize
                     space,               # the bounds on each dimension of x
                     acq_func="EI",       # the acquisition function
                     n_calls=100,         # the number of evaluations of f
                     n_random_starts=10,  # the number of random initialization points
                     random_state=0)

# Regret visualization
regret = [abs(x) for x in res_gp.func_vals]
plt.figure(figsize=(10, 5))
plt.plot(regret, label='Regret')
plt.xlabel('Iteration')
plt.ylabel('Absolute Regret')
plt.title('Regret of Bayesian Optimization')
plt.legend()
plt.show()
```
